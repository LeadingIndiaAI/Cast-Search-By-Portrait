```python
import os
import face_recognition
import matplotlib.pyplot as plt
from PIL import Image 
import cv2
import numpy as np
import pandas as pd
```

```python
images = os.listdir('C:/Users/Anil/Desktop/Cast_Search/test/candidates/')
#print(images)
test_path = 'C:/Users/Anil/Desktop/Cast_Search/test/candidates/'
```

```python
image = os.listdir('C:/Users/Anil/Desktop/Cast_Search/Face_Recog/')
path = 'C:/Users/Anil/Desktop/Cast_Search/Face_Recog/'+image[2]
#print(path)
```

```python
image_to_be_matched = face_recognition.load_image_file(path)
```

```python
image_to_be_matched_encoded = face_recognition.face_encodings(image_to_be_matched)[0]
```

```python
newpath = 'C:/Users/Anil/Desktop/Cast_Search/output' 
if not os.path.exists(newpath):
    os.makedirs(newpath)
```

```python
for im in images:
    current_image = face_recognition.load_image_file(test_path + im)
    encoding = face_recognition.face_encodings(current_image)
    if len(encoding) > 0:
        current_image_encoded = encoding[0]
    else:
        #print("No face detected ")
        continue
    result = face_recognition.compare_faces([image_to_be_matched_encoded], current_image_encoded)
    if result[0] == True:
        print("Detected: " + im)
        #output += image
        img_path = test_path + im
        img = Image.open(img_path)
        img.save(newpath + '/' + im)
    else:
        print("Not Detected : " + im)
```
